To build a machine learning system that can predict whether a credit card transection is legit or fraud.

# Work Flow

* Importing Libraries
* Importing Dataset
* Data Preprocessing
* Data Analysis
* Training and Testing data splitting 
* Model training (Logistic Regression)

# Importing Libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

# Importing Dataset

Link to dataset : https://www.kaggle.com/datasets/mlg-ulb/creditcardfraud

```python
data = pd.read_csv('creditcard.csv')
```

The data is highly unbalanced. More than 99% of the data belongs to one of the two classes.
* 0 : Legit transection
* 1 : Fraud transection

# Separating data for analysis

```python
legit = data[data.Class == 0]
fraud = data[data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

#### Statistical measure of the data

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

#### Compare the values for both transections

```python
data.groupby('Class').mean()
```

# Under-Sampling

Building a sample dataset containing similar distribution of normal transections and fraudulent transections.

#### Number of Fraudulent transections = 492

```python
legit_sample = legit.sample(n=492)
legit_sample
```

# Concatinating two dataframes to form new sample dataset

```python
new_data = pd.concat([legit_sample, fraud], axis=0)
new_data.head()
```

```python
new_data['Class'].value_counts()
```

```python
new_data.groupby('Class').mean()
```

# Splitting data into Features and Targets

```python
X = new_data.drop(columns='Class', axis=1)
Y = new_data['Class']
```

```python
X
```

```python
Y
```

# Splitting dataset into Training and Testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
print(Y.shape, Y_train.shape, Y_test.shape)
```

# Model Training

```python
model = LogisticRegression(n_jobs=-1)
```

```python
model.fit(X_train, Y_train)
```

# Model Evaluation

#### Accuracy on training data

```python
train_prediction = model.predict(X_train)
acc = accuracy_score(train_prediction, Y_train)
print("Accuracy of model on Training data is", acc*100,'%')
```

#### Accuracy on testing data

```python
test_prediction = model.predict(X_test)
acc2 = accuracy_score(test_prediction, Y_test)
print("Accuracy of model on testing data is", acc2*100)
```

# Building a Prediction System

```python
cols = list(X.columns)
li = []
for i in cols:
    x = float(input(f'Enter the value of {i} : '))
    li.append(x)

list_to_numpy_array = np.asarray(li)
reshaped_input = list_to_numpy_array.reshape(1,-1)
prediction = model.predict(reshaped_input)

if prediction[0] == 1:
    print("This is a FRAUDULANT TRANSECTION")
else:
    print("This is a LEGIT TRANSECTION")
```
